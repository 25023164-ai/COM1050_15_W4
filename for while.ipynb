```python
a = int(input(' nhap so nguyen a '))
b = int(input(' nhap so nguyen b '))
tong = 0
for i in range(a, b + 1):
    if i < 2:
        continue
    is_prime = True
    for n in range(2, int(i**0.5) + 1):
        if i % n == 0:
            is_prime = False
            break
    if is_prime:
            tong += i
print(' tong cac so nguyen to trong doan [a,b] là:', tong)        
    
    
```
